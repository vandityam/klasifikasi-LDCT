```python
import pandas as pd
import numpy as np
```

```python
# =====================================================
# 1. KLASIFIKASI SKOR TOTAL GURU
# =====================================================

# Load data guru
df_guru = pd.read_csv("files/data_LD_guru.csv", sep=';')
print("Jumlah data guru:", len(df_guru))

# Deteksi kolom Likert (skala 1-5)
likert_cols = [col for col in df_guru.columns if pd.api.types.is_numeric_dtype(df_guru[col])
               and df_guru[col].min() >= 1 and df_guru[col].max() <= 5]
non_likert_cols = [col for col in df_guru.columns if col not in likert_cols]

print(f"\nKolom Likert ({len(likert_cols)}): {likert_cols}")
print(f"Kolom Non-Likert ({len(non_likert_cols)}): {non_likert_cols}")
```

```python
# Ambil hanya kolom Likert
df_likert_guru = df_guru[likert_cols]
print("\nData Likert Guru:")
df_likert_guru.head(3)
```

```python
# Bagi dimensi
dim_skill = likert_cols[0:18]
dim_etika = likert_cols[18:25]
dim_keamanan = likert_cols[25:33]
dim_budaya = likert_cols[33:40]

# Hitung rata-rata per dimensi
df_guru["Skill_Digital"] = df_likert_guru[dim_skill].mean(axis=1)
df_guru["Etika_Digital"] = df_likert_guru[dim_etika].mean(axis=1)
df_guru["Keamanan_Digital"] = df_likert_guru[dim_keamanan].mean(axis=1)
df_guru["Budaya_Digital"] = df_likert_guru[dim_budaya].mean(axis=1)

print("\nSkor Guru per Dimensi:")
df_guru[["Skill_Digital", "Etika_Digital", "Keamanan_Digital", "Budaya_Digital"]].head()
```

```python
# 4. HITUNG SKOR TOTAL DAN RATA-RATA LD GURU
df_guru["Total_LD"] = df_guru[["Skill_Digital","Etika_Digital","Keamanan_Digital","Budaya_Digital"]].sum(axis=1)
df_guru["Mean_LD"] = df_guru[["Skill_Digital","Etika_Digital","Keamanan_Digital","Budaya_Digital"]].mean(axis=1)
print("\nRata-rata Skor Total Literasi Digital Guru:")
df_guru[["Total_LD", "Mean_LD"]].head()
```

```python
# 5. KLASIFIKASI SKOR TOTAL
# Klasifikasi guru
def kategori_ld(skor):
    if skor < 2.34:
        return "Rendah"
    elif skor < 3.67:
        return "Sedang"
    else:
        return "Tinggi"

df_guru["Level_LD"] = df_guru["Mean_LD"].apply(kategori_ld)

print("\nKlasifikasi Level Literasi Digital Guru:")
df_guru[["Mean_LD", "Level_LD"]].head()
```

```python
# =====================================================
# 2. KLASIFIKASI SKOR TOTAL SISWA
# =====================================================

# Load data siswa
df_siswa = pd.read_csv("files/data_ct_siswa.csv", sep=';')
print("\nJumlah data siswa:", len(df_siswa))

# Ambil kolom soal CT
ct_cols = [col for col in df_siswa.columns if col.startswith("S ")]
for col in ct_cols:
    df_siswa[col] = pd.to_numeric(df_siswa[col].astype(str).str.replace(",", "."), errors="coerce").fillna(0)

print("\nSoal_CT:", ct_cols)
```

```python
# 2. HITUNG SKOR TOTAL DAN RATA-RATA CT SISWA
df_siswa["Total_CT"] = df_siswa[ct_cols].sum(axis=1)
df_siswa["Mean_CT"] = df_siswa[ct_cols].mean(axis=1)
print("\nRata-rata Skor Total CT Siswa:")
df_siswa[["Total_CT", "Mean_CT"]].head()
```

```python
# 3. KLASIFIKASI SKOR TOTAL SISWA

soal_cols_per_jenjang = {
    "Siaga": [col for col in ct_cols if df_siswa.loc[df_siswa["Kategori"]=="Siaga", col].notna().any()],
    "Penggalang": [col for col in ct_cols if df_siswa.loc[df_siswa["Kategori"]=="Penggalang", col].notna().any()],
    "Penegak": [col for col in ct_cols if df_siswa.loc[df_siswa["Kategori"]=="Penegak", col].notna().any()]
}

def skor_max(row):
    jenjang = row["Kategori"]
    if jenjang == "Siaga":
        return len(soal_cols_per_jenjang["Siaga"]) * 8.33
    else:
        return len(soal_cols_per_jenjang[jenjang]) * 6.67

df_siswa["Persentase_CT"] = df_siswa.apply(lambda row: row["Total_CT"]/skor_max(row), axis=1)

# Klasifikasi siswa berdasarkan persentil per jenjang
def kategori_ct(df, skor_col="Persentase_CT", jenjang_col="Kategori"):
    df["Level_CT"] = ""
    for jenjang in df[jenjang_col].unique():
        mask = df[jenjang_col] == jenjang
        q1 = df.loc[mask, skor_col].quantile(0.33)
        q2 = df.loc[mask, skor_col].quantile(0.66)
        df.loc[mask, "Level_CT"] = df.loc[mask, skor_col].apply(lambda x: "Rendah" if x<=q1 else ("Sedang" if x<=q2 else "Tinggi"))
    return df

df_siswa = kategori_ct(df_siswa)

tampil_siswa = df_siswa.groupby("Kategori").head(3)
print("\nKlasifikasi Level CT Siswa:")
tampil_siswa[["Kategori", "Persentase_CT", "Level_CT"]].head(10)
```

```python
# =====================================================
# 3. PERBANDINGAN GURU & SISWA BERDASARKAN SEKOLAH
# =====================================================

def key_school(name):
    if pd.isna(name):
        return ""
    return " ".join(str(name).lower().strip().split()[:3])  

df_guru['sekolah_key'] = df_guru['Asal Instansi'].apply(key_school)
df_siswa['sekolah_key'] = df_siswa['SekolahNama'].apply(key_school)
```

```python
# Guru per sekolah
guru_per_sekolah = df_guru.groupby('sekolah_key').agg({
    'Asal Instansi':'first',
    'Mean_LD':'mean',
    'Level_LD': lambda x: x.mode()[0]
}).reset_index()

print("\nRata-rata Skor LD per Sekolah:")
guru_per_sekolah[["Asal Instansi", "Mean_LD", "Level_LD"]].head()
```

```python
# Siswa per sekolah
siswa_per_sekolah = df_siswa.groupby(['sekolah_key','SekolahNama']).agg({
    "Mean_CT":"mean",
    "Level_CT": lambda x: x.mode()[0]
}).reset_index()

print("\nRata-rata Skor LD per Sekolah:")
siswa_per_sekolah[["SekolahNama", "Mean_CT", "Level_CT"]].head()
```

```python
# Merge guru & siswa
df_perbandingan = guru_per_sekolah.merge(
    siswa_per_sekolah,
    on="sekolah_key",
    how="inner"
)[['Asal Instansi','SekolahNama','Level_LD','Level_CT','Mean_LD','Mean_CT']]

print("\nPerbandingan LD Guru & CT Siswa:")
df_perbandingan.head(10)
```

```python
from scipy.stats import pearsonr

# Hitung korelasi Pearson + p-value
r, p_value = pearsonr(df_perbandingan['Mean_LD'], df_perbandingan['Mean_CT'])
print("Korelasi (r) :", r)
print("P-value korelasi:", p_value)
```

Dari data yang ada, guru punya literasi digital tinggi, biasanya siswanya punya skor CT yang lebih tinggi. Hasil korelasinya cukup tinggi (r = 0.77), cuma secara statistik belum bisa dibilang signifikan (p = 0.13), kemungkinan karena jumlah sekolah yang dianalisis masih sedikit.

```python
# Pastikan guru per sekolah cuma 1 baris
df_guru_unique = df_guru.groupby('sekolah_key').agg({
    'Level_LD':'first'
}).reset_index()

# Merge dengan siswa
df_perbandingan = df_siswa.merge(
    df_guru_unique, 
    on='sekolah_key',
    how='left'
)

# Lihat distribusi per level guru
print("\nDistribusi Level Literasi Digital Guru:\n")
print(df_perbandingan['Level_LD'].value_counts())

tinggi = df_perbandingan[df_perbandingan['Level_LD'] == 'Tinggi']
sedang = df_perbandingan[df_perbandingan['Level_LD'] == 'Sedang']
rendah = df_perbandingan[df_perbandingan['Level_LD'] == 'Rendah']

print("\n=== Guru Tinggi ===")
print(tinggi['Level_CT'].value_counts())

print("\n=== Guru Sedang ===")
print(sedang['Level_CT'].value_counts())

print("\n=== Guru Rendah ===")
print(rendah['Level_CT'].value_counts())
```
